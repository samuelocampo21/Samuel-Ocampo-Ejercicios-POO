```python
#ejercicio 4#
class Edades:
  @staticmethod
  def edadal(juan):
    edadal=(2*juan)/3
    return edadal
  @staticmethod
  def edadana(juan):
    edadana=(4*juan)/3
    return edadana
  @staticmethod
  def edadmama(edadana,edadal,juan):
    edadmama=edadana+edadal+juan
    return edadmama
juan=9.0
print(f"La edad de Juan es: {juan}")
print(f"La edad de Alberto es: {Edades.edadal(juan)}")
print(f"La edad de Ana es: {Edades.edadana(juan)}")
print(f"La edad de la mamá es: {Edades.edadana(juan)}")
```

```python
#ejercicio 5#
class operaciones:
  @staticmethod
  def op (x,y):
    return x+y**2
suma=0
x=20
y=40
suma +=x
x = operaciones.op(x,y)

suma += (x/y)
print(f"El valor de x es: {x}")
print(f"El valor de y es: {y}")
print(f"La suma es: {suma}")
```

```python
#Ejercicio 12#
class op_nomina:
  @staticmethod
  def salario_bruto(horas_trabajadas, valor_hora):
        return horas_trabajadas * valor_hora
  @staticmethod
  def valor_retefuente(por_retefuente, salario_bruto):
        return por_retefuente * salario_bruto
  @staticmethod
  def salario_neto(salario_bruto, valor_retefuente):
        return salario_bruto - valor_retefuente

horas_trabajadas = 48.0
valor_hora = 5000.0
retencion = 12.5
por_retefuente = retencion / 100


salario_bruto = op_nomina.salario_bruto(horas_trabajadas, valor_hora)
valor_retefuente = op_nomina.valor_retefuente(por_retefuente, salario_bruto)
salario_neto = op_nomina.salario_neto(salario_bruto, valor_retefuente)


print(f"El salario bruto es de: {salario_bruto}")
print(f"La retención en la Fuente es de: {valor_retefuente}")
print(f"El salario neto es de: {salario_neto}")
```

```python
#Ejercicio 14#
class operaciones:
  @staticmethod
  def cuadrado(x):
    return x**2
  @staticmethod
  def cubo(x):
    return x**3
x=float(input("Ingrese el número que desea modificiar: "))
print(f"El número era: {x}")
print(f"El cuadrado del número es: {operaciones.cuadrado(x)}")
print(f"El cubo del número es: {operaciones.cubo(x)}")
```

```python
import math
class geometria:
  @staticmethod
  def area_circulo(radio):
    return math.pi*(pow(radio,2))
  @staticmethod
  def longuitud_circulo(radio):
    return 2*math.pi*radio
radio=float(input("Ingrese el radio del círculo: "))
print(f"El área del círculo es: {round(geometria.area_circulo(radio),2)}")
print(f"La longitud del círculo es: {round(geometria.longuitud_circulo(radio),2)}")
```
