```python
#ejercicio 4.2#
import tkinter as tk
from tkinter import messagebox
import math

class Notas:
    def __init__(self):
        self.lista_notas = [0] * 5

    def calcular_promedio(self):
        suma = 0

        for nota in self.lista_notas:
            suma += nota

        return suma / len(self.lista_notas)

    def calcular_desviacion(self):
        promedio = self.calcular_promedio()
        suma = 0

        for nota in self.lista_notas:
            suma += (nota - promedio) ** 2

        return math.sqrt(suma / len(self.lista_notas))

    def calcular_menor(self):
        menor = self.lista_notas[0]

        for nota in self.lista_notas:
            if nota < menor:
                menor = nota

        return menor


    def calcular_mayor(self):
        mayor = self.lista_notas[0]

        for nota in self.lista_notas:
            if nota > mayor:
                mayor = nota

        return mayor


class VentanaPrincipal:
    def __init__(self):
        self.ventana = tk.Tk()
        self.ventana.title("Notas")
        self.ventana.geometry("350x400")
        self.ventana.resizable(False, False)

        tk.Label(self.ventana, text="Nota 1:").place(x=20, y=20)
        self.campo_nota1 = tk.Entry(self.ventana)
        self.campo_nota1.place(x=120, y=20)

        tk.Label(self.ventana, text="Nota 2:").place(x=20, y=60)
        self.campo_nota2 = tk.Entry(self.ventana)
        self.campo_nota2.place(x=120, y=60)

        tk.Label(self.ventana, text="Nota 3:").place(x=20, y=100)
        self.campo_nota3 = tk.Entry(self.ventana)
        self.campo_nota3.place(x=120, y=100)

        tk.Label(self.ventana, text="Nota 4:").place(x=20, y=140)
        self.campo_nota4 = tk.Entry(self.ventana)
        self.campo_nota4.place(x=120, y=140)

        tk.Label(self.ventana, text="Nota 5:").place(x=20, y=180)
        self.campo_nota5 = tk.Entry(self.ventana)
        self.campo_nota5.place(x=120, y=180)

        boton_calcular = tk.Button(
            self.ventana,
            text="Calcular",
            command=self.calcular
        )
        boton_calcular.place(x=50, y=230)

        boton_limpiar = tk.Button(
            self.ventana,
            text="Limpiar",
            command=self.limpiar
        )
        boton_limpiar.place(x=180, y=230)

        self.promedio_label = tk.Label(
            self.ventana,
            text="Promedio = "
        )
        self.promedio_label.place(x=20, y=280)

        self.desviacion_label = tk.Label(
            self.ventana,
            text="Desviación estándar = "
        )
        self.desviacion_label.place(x=20, y=310)

        self.mayor_label = tk.Label(
            self.ventana,
            text="Nota mayor = "
        )
        self.mayor_label.place(x=20, y=340)

        self.menor_label = tk.Label(
            self.ventana,
            text="Nota menor = "
        )
        self.menor_label.place(x=20, y=370)

        self.ventana.mainloop()

    def calcular(self):
        try:
            notas = Notas()


            notas.lista_notas[0] = float(self.campo_nota1.get())
            notas.lista_notas[1] = float(self.campo_nota2.get())
            notas.lista_notas[2] = float(self.campo_nota3.get())
            notas.lista_notas[3] = float(self.campo_nota4.get())
            notas.lista_notas[4] = float(self.campo_nota5.get())

            promedio = notas.calcular_promedio()
            desviacion = notas.calcular_desviacion()
            mayor = notas.calcular_mayor()
            menor = notas.calcular_menor()

            self.promedio_label.config(
                text=f"Promedio = {promedio:.2f}"
            )

            self.desviacion_label.config(
                text=f"Desviación estándar = {desviacion:.2f}"
            )

            self.mayor_label.config(
                text=f"Nota mayor = {mayor}"
            )

            self.menor_label.config(
                text=f"Nota menor = {menor}"
            )

        except ValueError:
            messagebox.showerror(
                "Error",
                "Ingrese valores numéricos válidos"
            )

    def limpiar(self):
        self.campo_nota1.delete(0, tk.END)
        self.campo_nota2.delete(0, tk.END)
        self.campo_nota3.delete(0, tk.END)
        self.campo_nota4.delete(0, tk.END)
        self.campo_nota5.delete(0, tk.END)

        self.promedio_label.config(text="Promedio = ")
        self.desviacion_label.config(text="Desviación estándar = ")
        self.mayor_label.config(text="Nota mayor = ")
        self.menor_label.config(text="Nota menor = ")


if __name__ == "__main__":
    VentanaPrincipal()
```

```python
#ejercicio 4.3#
import tkinter as tk
from tkinter import messagebox
import math


class FiguraGeometrica:
    def __init__(self):
        self.__volumen = 0.0
        self.__superficie = 0.0

    def set_volumen(self, volumen):
        self.__volumen = volumen

    def set_superficie(self, superficie):
        self.__superficie = superficie

    def get_volumen(self):
        return self.__volumen

    def get_superficie(self):
        return self.__superficie


class Cilindro(FiguraGeometrica):

    def __init__(self, radio, altura):
        super().__init__()
        self.radio = radio
        self.altura = altura
        self.set_volumen(self.calcular_volumen())
        self.set_superficie(self.calcular_superficie())

    def calcular_volumen(self):
        return math.PI * self.altura * math.pow(self.radio, 2.0) if hasattr(math, 'PI') else math.pi * self.altura * math.pow(self.radio, 2.0)

    def calcular_superficie(self):
        area_lado_a = 2.0 * (math.PI if hasattr(math, 'PI') else math.pi) * self.radio * self.altura
        area_lado_b = 2.0 * (math.PI if hasattr(math, 'PI') else math.pi) * math.pow(self.radio, 2.0)
        return area_lado_a + area_lado_b


class Esfera(FiguraGeometrica):

    def __init__(self, radio):
        super().__init__()
        self.radio = radio
        self.set_volumen(self.calcular_volumen())
        self.set_superficie(self.calcular_superficie())

    def calcular_volumen(self):
        pi_val = math.PI if hasattr(math, 'PI') else math.pi
        return 1.333 * pi_val * math.pow(self.radio, 3.0)

    def calcular_superficie(self):
        pi_val = math.PI if hasattr(math, 'PI') else math.pi
        return 4.0 * pi_val * math.pow(self.radio, 2.0)


class Piramide(FiguraGeometrica):
    def __init__(self, base, altura, apotema):
        super().__init__()
        self.base = base
        self.altura = altura
        self.apotema = apotema
        self.set_volumen(self.calcular_volumen())
        self.set_superficie(self.calcular_superficie())

    def calcular_volumen(self):
        return (math.pow(self.base, 2.0) * self.altura) / 3.0

    def calcular_superficie(self):
        area_base = math.pow(self.base, 2.0)
        area_lado = 2.0 * self.base * self.apotema
        return area_base + area_lado



class VentanaPrincipal:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Cálculo de Figuras Geométricas")
        self.root.geometry("320x250")
        self.root.resizable(False, False)


        self.root.eval('tk::PlaceWindow . center')

        tk.Label(self.root, text="Seleccione una figura:", font=("Arial", 12, "bold")).pack(pady=15)


        tk.Button(self.root, text="Cilindro", width=20, command=self.abrir_cilindro).pack(pady=8)
        tk.Button(self.root, text="Esfera", width=20, command=self.abrir_esfera).pack(pady=8)
        tk.Button(self.root, text="Pirámide", width=20, command=self.abrir_piramide).pack(pady=8)

        self.root.mainloop()

    def abrir_cilindro(self):
        VentanaCilindro(self.root)

    def abrir_esfera(self):
        VentanaEsfera(self.root)

    def abrir_piramide(self):
        VentanaPiramide(self.root)


class VentanaCilindro(tk.Toplevel):
    def __init__(self, parent):
        super().__init__(parent)
        self.title("Cilindro")
        self.geometry("280x230")
        self.resizable(False, False)
        self.grab_set()


        tk.Label(self, text="Radio (cm):").place(x=20, y=20, width=100, height=23)
        self.campo_radio = tk.Entry(self)
        self.campo_radio.place(x=130, y=20, width=120, height=23)

        tk.Label(self, text="Altura (cm):").place(x=20, y=50, width=100, height=23)
        self.campo_altura = tk.Entry(self)
        self.campo_altura.place(x=130, y=50, width=120, height=23)

        self.btn_calcular = tk.Button(self, text="Calcular", command=self.calcular)
        self.btn_calcular.place(x=100, y=90, width=80, height=25)

        self.lbl_volumen = tk.Label(self, text="Volumen (cm³): ", anchor="w")
        self.lbl_volumen.place(x=20, y=140, width=240, height=23)

        self.lbl_superficie = tk.Label(self, text="Superficie (cm²): ", anchor="w")
        self.lbl_superficie.place(x=20, y=170, width=240, height=23)

    def calcular(self):
        try:
            r = float(self.campo_radio.get())
            h = float(self.campo_altura.get())

            cilindro = Cilindro(r, h)

            self.lbl_volumen.config(text=f"Volumen (cm³): {cilindro.get_volumen():.2f}")
            self.lbl_superficie.config(text=f"Superficie (cm²): {cilindro.get_superficie():.2f}")
        except ValueError:
            messagebox.showerror("Error", "Error de campo o dato no numérico")


class VentanaEsfera(tk.Toplevel):
    def __init__(self, parent):
        super().__init__(parent)
        self.title("Esfera")
        self.geometry("280x200")
        self.resizable(False, False)
        self.grab_set()

        tk.Label(self, text="Radio (cm):").place(x=20, y=20, width=100, height=23)
        self.campo_radio = tk.Entry(self)
        self.campo_radio.place(x=130, y=20, width=120, height=23)

        self.btn_calcular = tk.Button(self, text="Calcular", command=self.calcular)
        self.btn_calcular.place(x=100, y=60, width=80, height=25)

        self.lbl_volumen = tk.Label(self, text="Volumen (cm³): ", anchor="w")
        self.lbl_volumen.place(x=20, y=110, width=240, height=23)

        self.lbl_superficie = tk.Label(self, text="Superficie (cm²): ", anchor="w")
        self.lbl_superficie.place(x=20, y=140, width=240, height=23)

    def calcular(self):
        try:
            r = float(self.campo_radio.get())

            esfera = Esfera(r)

            self.lbl_volumen.config(text=f"Volumen (cm³): {esfera.get_volumen():.2f}")
            self.lbl_superficie.config(text=f"Superficie (cm²): {esfera.get_superficie():.2f}")
        except ValueError:
            messagebox.showerror("Error", "Error de campo o dato no numérico")


class VentanaPiramide(tk.Toplevel):

    def __init__(self, parent):
        super().__init__(parent)
        self.title("Pirámide")
        self.geometry("280x260")
        self.resizable(False, False)
        self.grab_set()

        tk.Label(self, text="Base (cm):").place(x=20, y=20, width=100, height=23)
        self.campo_base = tk.Entry(self)
        self.campo_base.place(x=130, y=20, width=120, height=23)

        tk.Label(self, text="Altura (cm):").place(x=20, y=50, width=100, height=23)
        self.campo_altura = tk.Entry(self)
        self.campo_altura.place(x=130, y=50, width=120, height=23)

        tk.Label(self, text="Apotema (cm):").place(x=20, y=80, width=100, height=23)
        self.campo_apotema = tk.Entry(self)
        self.campo_apotema.place(x=130, y=80, width=120, height=23)

        self.btn_calcular = tk.Button(self, text="Calcular", command=self.calcular)
        self.btn_calcular.place(x=100, y=120, width=80, height=25)

        self.lbl_volumen = tk.Label(self, text="Volumen (cm³): ", anchor="w")
        self.lbl_volumen.place(x=20, y=170, width=240, height=23)

        self.lbl_superficie = tk.Label(self, text="Superficie (cm²): ", anchor="w")
        self.lbl_superficie.place(x=20, y=200, width=240, height=23)

    def calcular(self):
        try:
            b = float(self.campo_base.get())
            h = float(self.campo_altura.get())
            a = float(self.campo_apotema.get())

            piramide = Piramide(b, h, a)

            self.lbl_volumen.config(text=f"Volumen (cm³): {piramide.get_volumen():.2f}")
            self.lbl_superficie.config(text=f"Superficie (cm²): {piramide.get_superficie():.2f}")
        except ValueError:
            messagebox.showerror("Error", "Error de campo o dato no numérico")

if __name__ == "__main__":
    VentanaPrincipal()
```
